```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
crick_df = pd.read_csv('matches.csv')
```

```python
toss_counts = crick_df["toss_decision"].value_counts()
plt.figure(figsize=(8,8))
toss_counts.plot(kind="pie", autopct='%1.0f%%')
plt.title("Toss decisions")
plt.ylabel('')
plt.show()

```

```python
most_won_team = crick_df["winner"].value_counts().head(10)
plt.figure(figsize=(10,6))
most_won_team.plot(kind="barh")
plt.title("Most won teams")
plt.xlabel("Number of wins")
plt.ylabel("Teams")
# plt.xticks(rotation = 45)
plt.show()
```

```python
winbyruns = (crick_df["win_by_runs"]!=0).sum()
winbywickets = (crick_df["win_by_wickets"]!=0).sum()
win_types = {'Runs' : winbyruns, 'Wickets': winbywickets}
plt.figure(figsize=(8,6))
plt.bar(win_types.keys(), win_types.values(), color = ['blue','orange'])
plt.title('Match win: Runs vs Wickets')
plt.xlabel('Win Type')
plt.ylabel('Number of matches')
plt.show()
```

```python
res = crick_df["result"].value_counts()
print(res)
```

```python
most_matches_venue = crick_df["venue"].value_counts().head(10)
plt.figure(figsize=(12,8))
most_matches_venue.plot(kind="barh")
plt.title("Top 10 Venues with most matches")
plt.xlabel("Number of matches hosted")
plt.ylabel("Venues")
plt.tight_layout()
plt.show()
```

```python
most_matches_city = crick_df["city"].value_counts()
print(most_matches_city.head(10))
```

```python
matches_per_season = crick_df["season"].value_counts().sort_index()
plt.figure(figsize=(12,6))
matches_per_season.plot(kind="line",marker = 'o')
plt.title("Cricket matches per season over time")
plt.xlabel('Season')
plt.ylabel('Number of matches')
plt.grid(True,alpha=0.5)
plt.show()
```
